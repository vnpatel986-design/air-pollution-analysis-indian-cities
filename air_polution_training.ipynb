```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

file_path = "/content/drive/MyDrive/50k.csv"
df = pd.read_csv(file_path)

print(df.shape)
print(df.columns)
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
if 'High_Pollution' not in df.columns:
    df['High_Pollution'] = df['US_AQI'].apply(
        lambda x: 1 if x >= 150 else 0
    )

print(df['High_Pollution'].value_counts())
```

```python
X = df.drop(columns=['High_Pollution'])
y = df['High_Pollution']

print(X.shape, y.shape)
```

```python
X_numeric = X.select_dtypes(include=['int64','float64'])
print(X_numeric.columns)
```

```python
from sklearn.feature_selection import VarianceThreshold

vt = VarianceThreshold(threshold=0.01)
X_var = vt.fit_transform(X_numeric)

cols = X_numeric.columns[vt.get_support()]
X_var_df = pd.DataFrame(X_var, columns=cols)

print(cols)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.impute import SimpleImputer
import pandas as pd

# Step 1: Impute missing values
imputer = SimpleImputer(strategy='mean')   # or 'median'
X_imputed = imputer.fit_transform(X_var_df)

# Step 2: Apply SelectKBest
skb = SelectKBest(score_func=f_classif, k=10)
X_kbest = skb.fit_transform(X_imputed, y)

# Step 3: Get selected column names
selected_cols = X_var_df.columns[skb.get_support()]
X_selected = pd.DataFrame(X_kbest, columns=selected_cols)

print(selected_cols)
```

```python
final_df = pd.concat(
    [X_selected.reset_index(drop=True),
     y.reset_index(drop=True)],
    axis=1
)

print(final_df.head())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(
    final_df.drop(columns=['High_Pollution'])
)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, final_df['High_Pollution'],
    test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
